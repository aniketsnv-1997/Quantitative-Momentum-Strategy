```python
from pandas_datareader import data
import pandas as pd

nifty = pd.read_csv("ind_nifty500list.csv")
company_symbol= nifty['Symbol'].tolist()
#print(company_symbol)

for i in company_symbol:
    try:
        df=data.DataReader(name=i, data_source="yahoo",start='2020-03-01',end='2021-02-28')
        open_start.append(df['Open'][0])
        open_end.append(df['Open'][-1])
        close_start.append(df['Close'][0])
        close_end.append(df['Close'][-1])
        tickers.append(i)
        std_end.append(df)
        #print(i)
    except:
        try:
            df=data.DataReader(name=i+".NS", data_source="yahoo",start='2020-03-01',end='2021-02-28')
            open_start.append(df['Open'][0])
            open_end.append(df['Open'][-1])
            close_start.append(df['Close'][0])
            close_end.append(df['Close'][-1])
            tickers.append(i+".NS")
            std_end.append(df)
            #print(i)
        except:
            open_start.append('NaN')
            open_end.append('NaN')
            close_start.append('NaN')
            close_end.append('NaN')
            tickers.append(i+".NS")
        #pass
    '''else:
        open_start.append('NaN')
        open_end.append('NaN')
        tickers.append(i)'''

#print(std_end)
#print(len(open_start), len(open_end))
#print(open_start)
#print(open_end)
```

```python
print(len(open_start),len(open_end),len(tickers),len(close_start),len(close_end))
#print(std_end[0]['Open'][0])
#print(std_end[0]['Open'][-1])

x = zip(open_start,open_end,tickers,close_start,close_end)
for i in x:
    print(i)
```

```python
nifty['Open_start']= open_start
nifty['Open_end']= open_end
nifty['Close_start']= close_start
nifty['Close_end']= close_end
```

```python
percentage_change=[]
for i, j in zip(open_start,open_end):
    percentage_change.append((float(j)-float(i))/float(i))
    
```

```python
nifty['Percentage_change']= percentage_change
nifty.to_csv('nifty_OC')
```
